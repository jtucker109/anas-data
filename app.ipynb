```python
audience['date'] = pd.to_datetime(audience['date'])
audience
```

```python
# audience[audience['date'].dt.year == 2024].groupby(audience['date'].dt.month)[['listeners','streams']].sum()
years = audience['date'].dt.year.unique()
dfs = []
for yr in years:
    monthly = audience[audience['date'].dt.year == yr].groupby(audience['date'].dt.month)[['listeners','streams']].sum()
    monthly['followers'] = audience[audience['date'].dt.year == yr].groupby(audience['date'].dt.month)['followers'].max()
    monthly['chg_followers'] = audience[audience['date'].dt.year == yr].groupby(audience['date'].dt.month)['followers'].max() - audience[audience['date'].dt.year == yr].groupby(audience['date'].dt.month)['followers'].min()
    monthly['year'] = yr
    dfs.append(monthly)

aud = pd.concat(dfs)
aud
```

```python
aud['month'] = pd.to_datetime(aud.index, format='%m')
aud['month'] = aud['month'].dt.month_name()
# aud.to_csv('monthly_data.csv',index=False)
aud
```

```python
aud24 = aud[aud['year'] == 2024]
aud25 = aud[aud['year'] == 2025]
aud26 = aud[aud['year'] == 2026]
aud24
```
